```python
import collections 
import re
import copy

DHSINFO_FIELDS = ['tissue', 'neqtl', 'ndhs', 'enrichment', 'pval']
class DhsInfo(collections.namedtuple('_DhsInfo', DHSINFO_FIELDS)):
    __slots__ = ()
    
CISINFO_FIELDS = ['tissue', 'ncis', 'ntrans', 'ncistrans', 'enrichment', 'pval']
class CisInfo(collections.namedtuple('_CisInfo', CISINFO_FIELDS)):
    __slots__ = ()

def read_cis_enrichments(filepath):
    res = dict()
    res2 = collections.defaultdict(dict)
    with open(filepath) as instream:
        for line in instream:
            if re.match("tissue", line):
                continue
            arr = line.strip().split("\t")
            tissue = arr[0]
            ncis   = int(arr[1])
            ntrans = int(arr[2])
            ncistrans  = int(arr[3])
            # randfrac = float(arr[4])
            enrich = float(arr[5])
            enrichment = np.log2(enrich) if enrich > 0 else 0.0
            pval = float(arr[6])
            res[tissue] = CisInfo(tissue=tissue, ncis=ncis, ntrans=ntrans, ncistrans=ncistrans, enrichment=enrichment, pval=pval)
            res2[tissue]['ncis'] = ncis
            res2[tissue]['ntrans'] = ntrans
            res2[tissue]['ncistrans']  = ncistrans
            res2[tissue]['enrichment'] = enrichment
            res2[tissue]['pval'] = pval
    return res, res2

def read_enrichments(filepath, ndhs_limit = 0):
    res = dict()
    res2 = collections.defaultdict(dict)
    with open(filepath) as instream:
        for line in instream:
            if re.match("tissue", line):
                continue
            arr = line.strip().split("\t")
            tissue = arr[0]
            neqtl = int(arr[3])
            ndhs  = int(arr[4])
            enrich = float(arr[5])
            enrichment = np.log2(enrich) if enrich > 0 else 0.0
            pval = float(arr[6])
            if ndhs > ndhs_limit:
                res[tissue] = DhsInfo(tissue=tissue, neqtl=neqtl, ndhs=ndhs, enrichment=enrichment, pval=pval)
                res2[tissue]['neqtl'] = neqtl
                res2[tissue]['ndhs']  = ndhs
                res2[tissue]['enrichment'] = enrichment
                res2[tissue]['pval'] = pval
            else:
                res[tissue] = DhsInfo(tissue=tissue, neqtl=neqtl, ndhs=ndhs, enrichment=0.0, pval=1.0)
                res2[tissue]['neqtl'] = neqtl
                res2[tissue]['ndhs']  = ndhs
                res2[tissue]['enrichment'] = 0.0
                res2[tissue]['pval'] = 1.0
    return res, res2

def read_enrichments_sb(filepath, ndhs_limit = 0):
    res = dict()
    res2 = collections.defaultdict(dict)
    with open(filepath) as instream:
        for line in instream:
            if re.match("tissue", line):
                continue
            arr = line.strip().split("\t")
            tissue = arr[0]
            neqtl = int(arr[2])
            ndhs  = int(arr[3])
            enrich = float(arr[4])
            enrichment = np.log2(enrich) if enrich > 0 else 0.0
            pval = float(arr[5])
            if ndhs > ndhs_limit:
                res[tissue] = DhsInfo(tissue=tissue, neqtl=neqtl, ndhs=ndhs, enrichment=enrichment, pval=pval)
                res2[tissue]['neqtl'] = neqtl
                res2[tissue]['ndhs']  = ndhs
                res2[tissue]['enrichment'] = enrichment
                res2[tissue]['pval'] = pval
            else:
                res[tissue] = DhsInfo(tissue=tissue, neqtl=neqtl, ndhs=ndhs, enrichment=0.0, pval=1.0)
                res2[tissue]['neqtl'] = neqtl
                res2[tissue]['ndhs']  = ndhs
                res2[tissue]['enrichment'] = 0.0
                res2[tissue]['pval'] = 1.0
    return res, res2

def read_enrichments_fs(filepath, ndhs_limit = 0):
    res = dict()
    res2 = collections.defaultdict(dict)
    with open(filepath) as instream:
        for line in instream:
            if re.match("tissue", line):
                continue
            arr = line.strip().split("\t")
            tissue = arr[0]
            neqtl = int(arr[1])
            ndhs  = int(arr[2])
            enrich = float(arr[3])
            enrichment = np.log2(enrich) if enrich > 0 else 0.0
            pval = float(arr[4])
            if ndhs > ndhs_limit:
                res[tissue] = DhsInfo(tissue=tissue, neqtl=neqtl, ndhs=ndhs, enrichment=enrichment, pval=pval)
                res2[tissue]['neqtl'] = neqtl
                res2[tissue]['ndhs']  = ndhs
                res2[tissue]['enrichment'] = enrichment
                res2[tissue]['pval'] = pval
            else:
                res[tissue] = DhsInfo(tissue=tissue, neqtl=neqtl, ndhs=ndhs, enrichment=0.0, pval=1.0)
                res2[tissue]['neqtl'] = neqtl
                res2[tissue]['ndhs']  = ndhs
                res2[tissue]['enrichment'] = 0.0
                res2[tissue]['pval'] = 1.0
    return res, res2

basedir = "/cbscratch/franco/trans-eqtl/dev-pipeline/"
# path1 = basedir+"gtex_v8_pub_lncRNA/raw/summary_5e-08/dhs_enrichments/permnull_sb0.1_knn30_crossmap"
# path2 = basedir+"gtex_v8_lncRNA_freeze/raw/summary_5e-08/dhs_enrichments/permnull_sb0.1_knn30"
# path3 = basedir+"gtex_v8_pub_lncRNA_nogtknn/raw/summary_5e-08/dhs_enrichments/permnull_sb0.1_knn30_crossmap_nogtknn"
# path4 = basedir+"gtex_v8_pub_lncRNA_nogtknn/raw/summary_5e-08/dhs_enrichments/permnull_sb0.1_knn30_nogtknn"
# path5 = basedir+"gtex_v8_pub_lncRNA/raw_pub/summary_5e-08/dhs_enrichments/permnull_sb0.1_knn30_crossmap"
# res_ph_cm, res_ph_cm_dict = read_enrichments(os.path.join(path1, target_file)) 
# res_ph, res_ph_dict = read_enrichments(os.path.join(path2, target_file)) 
# res_ph_cm_nogtknn, res_ph_cm_nogtknn_dict = read_enrichments(os.path.join(path3, target_file)) 
# res_ph_nogtknn, res_ph_nogtknn_dict = read_enrichments(os.path.join(path4, target_file))
# res_pub_cm, res_pub_cm_dict = read_enrichments(os.path.join(path5, target_file))

## SHAPEIT 2 data
path_old  = basedir+"gtex_v8_SHAPEIT2/raw_wasp/summary_5e-08/"
path_sb01    = "/cbscratch/franco/trans-eqtl/protein_coding_lncRNA_gamma01_knn30_cut5e-8/"
path_sb0006  = "/cbscratch/franco/trans-eqtl/protein_coding_lncRNA_gamma0006_knn30_cut5e-8/"
path_pub = basedir+"gtex_v8_SHAPEIT2/raw_pub/summary_5e-08/"
path_raw = basedir+"gtex_v8_SHAPEIT2/raw/summary_5e-08/"


dhs_target_file = "dhs_enrichment_multi_tissue_5e-08.txt.ld_prune"
cis_target_file = "ciseqtl_enrichment.txt.ld_prune"

# uses saikat's format
dhs01       , dhs_dict01        = read_enrichments_sb(os.path.join(path_sb01,"dhs_enrichments/dhs_enrichment_multi_tissue_pc_lncRNA_gamma01_knn30_cut5e-8_ldpruned.txt")) 
dhs_cm01    , dhs_cm_dict01     = read_enrichments_fs(os.path.join(path_raw, "dhs_enrichments/permnull_sb0.1_knn30_crossmap", dhs_target_file))
dhs_nocis01 , dhs_nocis_dict01  = read_enrichments_fs(os.path.join(path_raw, "dhs_enrichments/permnull_sb0.1_knn30_nocismask", dhs_target_file))

dhs0006       , dhs_dict0006        = read_enrichments_sb(os.path.join(path_sb0006,"dhs_enrichments/dhs_enrichment_multi_tissue_pc_lncRNA_gamma0006_knn30_cut5e-8_ldpruned.txt")) 
dhs_cm0006    , dhs_cm_dict0006     = read_enrichments_fs(os.path.join(path_raw, "dhs_enrichments/permnull_sb0.006_knn30_crossmap", dhs_target_file))
dhs_nocis0006 , dhs_nocis_dict0006  = read_enrichments_fs(os.path.join(path_raw, "dhs_enrichments/permnull_sb0.006_knn30_nocismask", dhs_target_file))


### Outdated
# dhs_wasp  , dhs_wasp_dict   = read_enrichments(os.path.join(path_old, "dhs_enrichments/permnull_sb0.1_knn30", dhs_target_file))
# dhs_pub_cm, dhs_pub_cm_dict = read_enrichments_new_less(os.path.join(path_pub, "dhs_enrichments/permnull_sb0.1_knn30_crossmap", dhs_target_file))
# dhs_pub   , dhs_pub_dict    = read_enrichments_new_less(os.path.join(path_pub, "dhs_enrichments/permnull_sb0.1_knn30", dhs_target_file))


### Load cis Enrichments
cis01       , cis_dict01        = read_cis_enrichments(os.path.join(path_sb01,"GTExPortal_eqtl_analysis/ciseqtl_enrichment_pc_lncRNA_gamma01_knn30_cut5e-8_ldpruned.txt")) 
cis_nocis01 , cis_nocis_dict01  = read_cis_enrichments(os.path.join(path_raw, "GTExPortal_eqtl_analysis/permnull_sb0.1_knn30_nocismask", cis_target_file))
cis_cm01    , cis_cm_dict01     = read_cis_enrichments(os.path.join(path_raw, "GTExPortal_eqtl_analysis/permnull_sb0.1_knn30_crossmap", cis_target_file))

cis0006       , cis_dict0006        = read_cis_enrichments(os.path.join(path_sb0006,"GTExPortal_eqtl_analysis/ciseqtl_enrichment_pc_lncRNA_gamma0006_knn30_cut5e-8_ldpruned.txt")) 
cis_nocis0006 , cis_nocis_dict0006  = read_cis_enrichments(os.path.join(path_raw, "GTExPortal_eqtl_analysis/permnull_sb0.006_knn30_nocismask", cis_target_file))
cis_cm0006    , cis_cm_dict0006     = read_cis_enrichments(os.path.join(path_raw, "GTExPortal_eqtl_analysis/permnull_sb0.006_knn30_crossmap", cis_target_file))

### Outdated
# cis_pub_cm, cis_pub_cm_dict = read_cis_enrichments(os.path.join(path_pub, "GTExPortal_eqtl_analysis/permnull_sb0.1_knn30_crossmap", cis_target_file))
# cis_pub   , cis_pub_dict    = read_cis_enrichments(os.path.join(path_pub, "GTExPortal_eqtl_analysis/permnull_sb0.1_knn30", cis_target_file))
```

```python
import mpmath
mpmath.mp.dps = 50
def pvalue(x): return float(mpmath.log10(1 - 0.5 * (1 + mpmath.erf(x/mpmath.sqrt(2)))))

SNPRES_FIELDS = ['rsid', 'chrom', 'pos', 'logp', 'maf']
class SNPRes(collections.namedtuple('_SNPRes', SNPRES_FIELDS)):
    __slots__ = ()
    
def tejaas(filepath):
    res = list()
    with open(filepath, 'r') as mfile:
        next(mfile)
        for line in mfile:
            arr   = line.strip().split("\t")
            rsid  = arr[0]
            chrom = int(arr[1])
            pos   = int(arr[2])
            maf   = float(arr[3])
            q     = float(arr[4])
            mu    = float(arr[5])
            sigma = float(arr[6])
            p     = float(arr[7])
            if sigma == 0:
                continue
            logp  = np.log10(p) if p != 0 else pvalue( (q - mu) / sigma)
            res.append(SNPRes(rsid=rsid, chrom=chrom, pos=pos, logp=-logp, maf=maf))
    return res

## SHAPEIT 2 data
path_old  = basedir+"gtex_v8_SHAPEIT2/raw_wasp/summary_5e-08/"
# path_sb  = "/cbscratch/franco/trans-eqtl/protein_coding_lncRNA_gamma01_knn30_cut5e-8/"
# alt_path_sb  = "/cbscratch/franco/trans-eqtl/protein_coding_lncRNA_gamma0006_knn30_cut5e-8/"
path_sb01    = "/cbscratch/franco/trans-eqtl/protein_coding_lncRNA_gamma01_knn30_cut5e-8/"
path_sb0006  = "/cbscratch/franco/trans-eqtl/protein_coding_lncRNA_gamma0006_knn30_cut5e-8/"

path_pub = basedir+"gtex_v8_SHAPEIT2/raw_pub/summary_5e-08/"
path_raw = basedir+"gtex_v8_SHAPEIT2/raw/summary_5e-08/"


target_file    = "trans_eqtls.txt.ld_prune"

teqtls_wasp   = os.path.join(path_old, "{:s}/tejaas", "permnull_sb0.1_knn30", target_file)
# uses saikat's format
teqtls        = os.path.join(path_sb01, "{:s}/trans_eqtls_ldpruned.txt") 
teqtls_cm     = os.path.join(path_raw, "{:s}/tejaas", "permnull_sb0.1_knn30_crossmap", target_file)
teqtls_nocis  = os.path.join(path_raw, "{:s}/tejaas", "permnull_sb0.1_knn30_nocismask", target_file)


alt_teqtls        = os.path.join(path_sb0006, "{:s}/trans_eqtls_ldpruned.txt") 
alt_teqtls_cm     = os.path.join(path_raw, "{:s}/tejaas", "permnull_sb0.006_knn30_crossmap", target_file)
alt_teqtls_nocis  = os.path.join(path_raw, "{:s}/tejaas", "permnull_sb0.006_knn30_nocismask", target_file)

teqtls_pub_cm = os.path.join(path_pub, "{:s}/tejaas", "permnull_sb0.1_knn30_crossmap", target_file)
teqtls_pub    = os.path.join(path_pub, "{:s}/tejaas", "permnull_sb0.1_knn30", target_file)


teqtl_dict = collections.defaultdict(dict)
for ts in tshorts:
    if ts not in brain_tissues:
        if ts in optim_tissues:
            teqtl_dict["sb"][ts] = tejaas(alt_teqtls.format(ts))
            teqtl_dict["cm"][ts] = tejaas(alt_teqtls_cm.format(ts))
            teqtl_dict["nocis"][ts] = tejaas(alt_teqtls_nocis.format(ts))
        else:
            teqtl_dict["sb"][ts] = tejaas(teqtls.format(ts))
            teqtl_dict["cm"][ts] = tejaas(teqtls_cm.format(ts))
            teqtl_dict["nocis"][ts] = tejaas(teqtls_nocis.format(ts))
            teqtl_dict["pub_cm"][ts] = tejaas(teqtls_pub_cm.format(ts))
            teqtl_dict["pub"][ts] = tejaas(teqtls_pub.format(ts))
```

```python
unique_set = collections.defaultdict(set)
for ts in tshorts:
    if ts not in brain_tissues:
        if ts in optim_tissues:
            unique_set["sb"].update(set([x.rsid for x in teqtl_dict["sb"][ts]]))
            unique_set["cm"].update(set([x.rsid for x in teqtl_dict["cm"][ts]]))
            unique_set["nocis"].update(set([x.rsid for x in teqtl_dict["nocis"][ts]]))
        else:
            unique_set["sb"].update(set([x.rsid for x in teqtl_dict["sb"][ts]]))
            unique_set["cm"].update(set([x.rsid for x in teqtl_dict["cm"][ts]]))
            unique_set["nocis"].update(set([x.rsid for x in teqtl_dict["nocis"][ts]]))
            unique_set["pub_cm"].update(set([x.rsid for x in teqtl_dict["pub_cm"][ts]]))
            unique_set["pub"].update(set([x.rsid for x in teqtl_dict["pub"][ts]]))
```
