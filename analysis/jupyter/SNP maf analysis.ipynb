```python
%load_ext autoreload
%autoreload 2
import numpy as np
import matplotlib.pyplot as plt
import os
import sys
sys.path.append('../')
sys.path.append('/usr/users/fsimone/tejaas')
import os
from utils import utils

snpinfo_file = "/cbscratch/franco/datasets/gtex_v8/genotypes/vcfs_0.01/gtex_v8_snpinfo.txt"
snp_dict = dict()
with open(snpinfo_file) as instream:
    for line in instream:
        arr = line.rstrip().split()
        maf = float(arr[5])
        varid = arr[1]
        snp_dict[varid] = maf
```

```python
import collections
tissue_file = "/usr/users/fsimone/trans-eqtl-pipeline/main/tissues.txt"
tissuenames, descriptions = utils.read_tissues(tissue_file)

teqtl_dict = collections.defaultdict(dict)
for t in tissuenames:
    print(t)
    infile = "/cbscratch/franco/trans-eqtl/dev-pipeline/gtex_v8_tejaas_permnull_sb0.1_knn/{:s}/trans_eqtls.txt".format(t)
    if os.path.exists(infile):
        with open(infile) as instream:
            next(instream)
            for line in instream:
                arr = line.rstrip().split()
                teqtl_dict[t][arr[0]] = float(arr[3])
    
```

```python
all_maf = list()
all005_maf = list()
labels = list()
maf_cutoff = 0.05
print("{: >10}|{: >10}|{: >10}|{: >10}|{: >10}".format("Tissue", "maf<0.01", "maf>0.01", "maf>0.05", "diff"))
for t in tissuenames:
    if t in teqtl_dict:
        labels.append(t)
        missing_snps = 0
        maf_values = list()
        maf005_values = list()
        for x in list(teqtl_dict[t].keys()):
            if snp_dict.get(x, False):
                maf_values.append(snp_dict[x])
                if snp_dict[x] >= maf_cutoff and snp_dict[x] <= (1-maf_cutoff):
                    maf005_values.append(snp_dict[x])
            else:
                missing_snps += 1
        all_maf.append(maf_values)
        all005_maf.append(maf005_values)
        print("{: >10}|{: >10}|{: >10}|{: >10}|{: >10}".format(t, missing_snps, len(maf_values), len(maf005_values), len(maf_values) - len(maf005_values)))
```
