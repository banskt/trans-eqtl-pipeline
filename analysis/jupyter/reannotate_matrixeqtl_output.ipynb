```python
gtposfile = "/cbscratch/franco/datasets/gtex/genotypes/GTEx_v6p_all_SNPs_pos_filtered.txt"
mqtldir  = "/cbscratch/franco/trans-eqtl/dev-pipeline/gtex_v6_gtknn/raw/gtex-ms/matrixeqtl/chr21/"

# load SNP rsid position dictionary
snpdict = dict()
with open(gtposfile) as instream:
    for line in instream:
        arr = line.rstrip().split()
        snpdict[arr[1]] = arr[2]

mqtlfiles = ["trans_eqtls", "cis_eqtls"]

for mfile in mqtlfiles:
    with open(mqtldir+"/"+mfile+"_pos.txt", 'w') as outstream:
        with open(mqtldir+"/"+mfile+".txt") as instream:
            next(instream)
            for line in instream:
                newline = line.rstrip()
                rsid    = newline.split()[0]
                if snpdict.get(rsid, False):
                    outstream.write(newline+"\t"+snpdict[rsid]+"\n")
                else:
                    print("ERROR: RSID not found")
                    raise
        
```
