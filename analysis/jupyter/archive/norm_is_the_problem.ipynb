```python
dim = 2000
```

```python
mcov = np.eye(dim)
```

```python
mmean = np.repeat(0, dim)
```

```python
msize = (2)
X = np.random.multivariate_normal(mmean, mcov, msize)
```

```python
X.shape
```

```python
np.linalg.norm(X, axis = 0).shape
```

```python
Xprime = X / np.linalg.norm(X, axis = 0)
```

```python
np.linalg.norm(Xprime, axis = 0).shape
```

```python
Xprime.shape
```

```python
fig = plt.figure(figsize = (14, 6))
ax1 = fig.add_subplot(121)
ax2 = fig.add_subplot(122)
ax1.hist(Xprime[0, :])
ax2.hist(Xprime[1, :])
plt.show()
```
